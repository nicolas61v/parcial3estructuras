## 1. M茅todo de Simpson 1/3 Simple

Para empezar con algo relativamente sencillo, implementamos el m茅todo de Simpson 1/3 simple. Este m茅todo b谩sicamente toma tres puntos: los extremos del intervalo y el punto medio.

### La teor铆a (breve)
La f贸rmula que usamos es:

$$\int_a^b f(x)dx \approx \frac{h}{3}[f(a) + 4f(\frac{a+b}{2}) + f(b)]$$

donde $h = \frac{b-a}{2}$ (es el ancho del intervalo).

### El c贸digo
Aqu铆 est谩 la implementaci贸n:

```c
// Primero se define la funci贸n que vamos a integrar (la que nos dio en el ejercicio)
double f(double x) {
    return exp(-x * x);  // f(x) = e^(-x^2)
}

// Versi贸n normal (secuencial)
double simpson_one_third_simple(double a, double b) {
    double h = (b - a) / 2;
    return (h / 3) * (f(a) + 4 * f(a + h) + f(b));
}

// Versi贸n paralela (con OpenMP)
double simpson_one_third_simple_parallel(double a, double b) {
    double h = (b - a) / 2;
    double sum_middle;
    
    #pragma omp parallel shared(h)
    {
        #pragma omp single
        sum_middle = f(a + h);
    }
    
    return (h / 3) * (f(a) + 4 * sum_middle + f(b));
}
```

## 2. M茅todo de Simpson 3/8 Simple

Este m茅todo es similar al anterior pero usa cuatro puntos en lugar de tres.

### La teor铆a
La f贸rmula es un poco m谩s compleja:

$$\int_a^b f(x)dx \approx \frac{3h}{8}[f(a) + 3f(a+h) + 3f(a+2h) + f(b)]$$

donde $h = \frac{b-a}{3}$

### El c贸digo

```c
// Versi贸n normal
double simpson_three_eighths_simple(double a, double b) {
    double h = (b - a) / 3;
    return (3 * h / 8) * (f(a) + 3 * f(a + h) + 3 * f(a + 2 * h) + f(b));
}

// Versi贸n paralela
double simpson_three_eighths_simple_parallel(double a, double b) {
    double h = (b - a) / 3;
    double sum_middle = 0;
    
    #pragma omp parallel sections reduction(+:sum_middle)
    {
        #pragma omp section
        sum_middle += f(a + h);
        
        #pragma omp section
        sum_middle += f(a + 2 * h);
    }
    
    return (3 * h / 8) * (f(a) + 3 * sum_middle + f(b));
}
```

## 3. M茅todo de Simpson 1/3 Compuesto

Ahora s铆 viene lo interesante. Este m茅todo divide el intervalo en muchos pedazos y aplica la regla de Simpson 1/3 en cada uno.

### La teor铆a
La f贸rmula se ve intimidante pero no es tan complicada:

$$\int_a^b f(x)dx \approx \frac{h}{3}[f(a) + 4\sum_{i=1,\text{impar}}^{n-1}f(x_i) + 2\sum_{i=2,\text{par}}^{n-2}f(x_i) + f(b)]$$

### El c贸digo

```c
// Versi贸n normal
double simpson_one_third_composite(double a, double b, int n) {
    if (n % 2 != 0) n++; // Para asegurarnos que n es par
    double h = (b - a) / n;
    double sum_odd = 0.0, sum_even = 0.0;

    for (int i = 1; i <= n-1; i += 2) {
        sum_odd += f(a + i * h);
    }
    
    for (int i = 2; i <= n-2; i += 2) {
        sum_even += f(a + i * h);
    }

    return (h / 3) * (f(a) + 4 * sum_odd + 2 * sum_even + f(b));
}

// Versi贸n paralela
double simpson_one_third_composite_parallel(double a, double b, int n) {
    if (n % 2 != 0) n++;
    double h = (b - a) / n;
    double sum_odd = 0.0, sum_even = 0.0;

    #pragma omp parallel sections reduction(+:sum_odd,sum_even)
    {
        #pragma omp section
        {
            #pragma omp parallel for reduction(+:sum_odd)
            for (int i = 1; i <= n-1; i += 2) {
                sum_odd += f(a + i * h);
            }
        }

        #pragma omp section
        {
            #pragma omp parallel for reduction(+:sum_even)
            for (int i = 2; i <= n-2; i += 2) {
                sum_even += f(a + i * h);
            }
        }
    }

    return (h / 3) * (f(a) + 4 * sum_odd + 2 * sum_even + f(b));
}
```

## 4. M茅todo de Simpson 3/8 Compuesto

El 煤ltimo m茅todo! Este es similar al anterior pero usa la regla de Simpson 3/8 en cada subintervalo.

### La teor铆a

$$\int_a^b f(x)dx \approx \frac{3h}{8}[f(a) + 3\sum_{i=1}^{n-2} f(x_i) + 3\sum_{i=2}^{n-1} f(x_i) + f(b)]$$

### El c贸digo

```c
// Versi贸n normal
double simpson_three_eighths_composite(double a, double b, int n) {
    if (n % 3 != 0) n += (3 - n % 3); // Aseguramos que n sea m煤ltiplo de 3
    double h = (b - a) / n;
    double sum_1 = 0.0, sum_2 = 0.0;

    for (int i = 1; i <= n-2; i += 3) {
        sum_1 += f(a + i * h);
    }
    
    for (int i = 2; i <= n-1; i += 3) {
        sum_2 += f(a + i * h);
    }

    return (3 * h / 8) * (f(a) + 3 * sum_1 + 3 * sum_2 + f(b));
}

// Versi贸n paralela
double simpson_three_eighths_composite_parallel(double a, double b, int n) {
    if (n % 3 != 0) n += (3 - n % 3);
    double h = (b - a) / n;
    double sum_1 = 0.0, sum_2 = 0.0;

    #pragma omp parallel sections reduction(+:sum_1,sum_2)
    {
        #pragma omp section
        {
            #pragma omp parallel for reduction(+:sum_1)
            for (int i = 1; i <= n-2; i += 3) {
                sum_1 += f(a + i * h);
            }
        }

        #pragma omp section
        {
            #pragma omp parallel for reduction(+:sum_2)
            for (int i = 2; i <= n-1; i += 3) {
                sum_2 += f(a + i * h);
            }
        }
    }

    return (3 * h / 8) * (f(a) + 3 * sum_1 + 3 * sum_2 + f(b));
}
```
